```python
from lib.crypto.csprng import *
import matplotlib.pyplot as plt
from random import random, SystemRandom
from lib.crypto.aes import *
import numpy as np
```

## Python Secure Random (Float Mode)

```python
x = SystemRandom()
```

```python
value = []

for i in range(1_000_000):
  value.append(x.random())

plt.hist(value)
```

```python
plt.hist(to_linear(value))
```

```python
plt.hist(np.diff(value))
```

## Python Secure Random (Integer Mode)

```python
x = SystemRandom()
value = []

for i in range(1_000_000):
  value.append(x.randint(0, 255))

plt.hist(value)
```

```python
plt.hist(np.diff(value))
```

## Henon Map

```python
x = HenonMap(random(), random(), random())

value = []
for i in range(1_000_000):
  val = x.get_tuple()[0]
  value.append(val)

  x = x.next()

plt.hist(value)
```

```python
plt.hist(to_linear(value))
```

```python
plt.hist(np.diff(value))
```
